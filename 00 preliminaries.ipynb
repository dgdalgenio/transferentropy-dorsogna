# Check if required packages are complete

```python
import importlib
import subprocess

packages = {
    "numpy": "numpy",
    "matplotlib": "matplotlib",
    "scipy": "scipy",
    "pandas": "pandas",
    "sklearn": "scikit-learn",
    "tqdm": "tqdm",
    "pyinform": "pyinform",
    "jpype1": "jpype1",
}

for import_name, install_name in packages.items():
    if importlib.util.find_spec(import_name) is None:
        print(f"Installing {install_name}...")
        subprocess.run(["pip", "install", install_name], check=True)
    else:
        print(f"{install_name} already installed.")
```
